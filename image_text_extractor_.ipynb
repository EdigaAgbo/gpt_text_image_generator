```python
import PyPDF2
import fitz
```

```python
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
def extract_images_from_pdf(pdf_file, image_dir):
    doc = fitz.open(pdf_file)
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        image_list = page.get_images(full=True)
        for img_index, img in enumerate(image_list):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_data = base_image["image"]
            with open(f"{image_dir}/image_{page_num}_{img_index}.png", "wb") as image_file:
                image_file.write(image_data)
```

```python
if __name__ == "__main__":
    pdf_file = "/content/drive/MyDrive/Colab Notebooks/image_text_extraction_/pdf_file.pdf"
    text = extract_text_from_pdf(pdf_file)
    with open("/content/drive/MyDrive/Colab Notebooks/image_text_extraction_/extracted_text.txt", "w", encoding="utf-8") as text_file:
        text_file.write(text)

    image_dir = "/content/drive/MyDrive/Colab Notebooks/image_text_extraction_/extracted_images"
    extract_images_from_pdf(pdf_file, image_dir)
    print("Images extracted and saved in the 'image_extraction' directory.")
```
